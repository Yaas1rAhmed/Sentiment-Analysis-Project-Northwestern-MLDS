## N-Gram Tokenization

```python
pdf_files = ['062923_speech.pdf', '113022_speech.pdf', '082622_speech.pdf', '032122_speech.pdf', 
            '082721_speech.pdf', '050321_speech.pdf', '021021_speech.pdf', '100620_speech.pdf',
            '082720_speech.pdf', '051320_speech.pdf', '040920_speech.pdf']

# Create a SentimentIntensityAnalyzer instance
sid = SentimentIntensityAnalyzer()

# Extract text from PDF files
texts = []
for pdf_file in pdf_files:
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_number in range(len(reader.pages)):
            page = reader.pages[page_number]
            text += page.extract_text()
        texts.append(text)

# Define the value of n for n-gram tokenization
n = 7  # Choose the desired n-gram size

for i, text in enumerate(texts):
    words = nltk.word_tokenize(text)
    ngrams_list = list(ngrams(words, n))

    sentiment_scores = []

    for ngram in ngrams_list:
        ngram_text = ' '.join(ngram)
        ngram_scores = sid.polarity_scores(ngram_text)
        sentiment_scores.append(ngram_scores['compound'])

    avg_score = sum(sentiment_scores) / len(sentiment_scores)

    # Interpret the sentiment scores
    if avg_score >= 0.05:
        sentiment = 'Positive'
    elif avg_score <= -0.05:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'

    # Print the sentiment and average sentiment score for each speech
    print("Speech", i+1)
    print("Sentiment: ", sentiment)
    print("Average Sentiment Score: ", avg_score)
    print("--------------------------------------")
    
    
```

```python
pdf_files = ['062923_speech.pdf', '113022_speech.pdf', '082622_speech.pdf', '032122_speech.pdf', 
            '082721_speech.pdf', '050321_speech.pdf', '021021_speech.pdf', '100620_speech.pdf',
            '082720_speech.pdf', '051320_speech.pdf', '040920_speech.pdf']

# Define common stop words and punctuation characters
common_stop_words = set(stopwords.words('english'))
punctuations = set(string.punctuation)
additional_chars_to_exclude = set(['"', "'", '"'])

# Extract text from PDF files
speeches_tokens = []
for pdf_file in pdf_files:
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_number in range(len(reader.pages)):
            page = reader.pages[page_number]
            text += page.extract_text()

        # Tokenize and preprocess the text
        sentences = sent_tokenize(text)
        tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]

        # Flatten the list of tokenized words for the speech
        speech_tokens = [word.lower() for tokens in tokenized_sentences for word in tokens]

        # Filter out punctuation, common stop words, numbers, and additional characters
        speech_tokens = [word for word in speech_tokens if word not in common_stop_words
                         and word not in punctuations and word.isalpha() and word not in additional_chars_to_exclude]

        # Append the cleaned speech tokens to the list
        speeches_tokens.append(speech_tokens)

# Calculate the most common words for each speech
most_common_words_per_speech = []
for speech_tokens in speeches_tokens:
    word_counter = Counter(speech_tokens)
    most_common_words = word_counter.most_common(10)  # You can adjust the number of common words
    most_common_words_per_speech.append(most_common_words[:10])
```

```python
import yfinance as yf
import pandas as pd

sym1 = 'SPY'
sym2 = '^VIX'
start_dates = ['2020-04-01', '2020-05-07', '2020-08-20', '2020-10-01', '2021-02-03', '2021-04-27', '2021-08-20',
             '2022-03-10', '2022-08-20', '2022-11-25', '2023-06-15']

end_dates = ['2020-05-01', '2020-06-07', '2020-09-20', '2020-10-30', '2021-03-01', '2021-05-27', '2021-09-20',
           '2022-04-10', '2022-09-20', '2022-12-25', '2023-07-15']

spy_data_all = pd.DataFrame()
vix_data_all = pd.DataFrame()

# Loop through the start and end date pairs
for start_date, end_date in zip(start_dates, end_dates):
    spy_data = yf.download(sym1, start=start_date, end=end_date)
    vix_data = yf.download(sym2, start=start_date, end=end_date)
    
    # Add a 'Start Date' and 'End Date' column to the data
    spy_data['Start Date'] = start_date
    spy_data['End Date'] = end_date
    vix_data['Start Date'] = start_date
    vix_data['End Date'] = end_date
    
    # Concatenate the data to the existing DataFrame
    spy_data_all = pd.concat([spy_data_all, spy_data])
    vix_data_all = pd.concat([vix_data_all, vix_data])
```

```python
# Define the date range based on start_dates and end_dates
date_ranges = list(zip(start_dates, end_dates))
num_date_ranges = len(date_ranges)

# Define the specific dates for vertical lines
speech_dates = ['2020-04-09', '2020-05-13', '2020-08-27', '2020-10-06', '2021-02-10', '2021-05-03',
               '2021-08-27', '2022-03-21', '2022-08-26', '2022-11-30', '2023-06-29']

# Convert speech_dates to datetime objects
speech_dates = [pd.to_datetime(date) for date in speech_dates]

# Create the interactive slider
slider = widgets.IntSlider(min=0, max=num_date_ranges - 1, step=1, value=0, description="Date Range")

# Store the lines for each date
speech_lines = []

def update_plot(date_range_index):
    global speech_lines
    start_date, end_date = date_ranges[date_range_index]
    selected_spy_data = spy_data_all[(spy_data_all['Start Date'] == start_date) & (spy_data_all['End Date'] == end_date)]
    selected_vix_data = vix_data_all[(vix_data_all['Start Date'] == start_date) & (vix_data_all['End Date'] == end_date)]

    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    ax[0].plot(selected_spy_data.index, selected_spy_data['Close'])
    ax[0].tick_params(axis='x', rotation=90)
    ax[0].set_ylabel("Price")
    ax[0].set_title("S&P 500 Index ($SPY)")

    ax[1].plot(selected_vix_data.index, selected_vix_data['Close'])
    ax[1].tick_params(axis='x', rotation=90)
    ax[1].set_ylabel("Price")
    ax[1].set_title("Volatility Index ($VIX)")

    # Add vertical line on specific date in VIX plot
    speech_date = speech_dates[date_range_index]
    vline = ax[1].axvline(x=speech_date, color='r', linestyle='--', label='Speech Date')
    vline_spy = ax[0].axvline(x=speech_date, color='r', linestyle='--', label='Speech Date')
    speech_lines.append(vline)

    ax[1].legend()
    plt.tight_layout()
    for a in ax:
        sns.despine(ax = a)
    plt.show()

# Create the interactive plot using interactive_output
interactive_plot = interactive_output(update_plot, {'date_range_index': slider})

# Display the widgets and the interactive plot
display(widgets.VBox([slider, interactive_plot]))
```
